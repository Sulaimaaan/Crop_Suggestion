```python
df=pd.read_csv('Crop_recommendation.csv')
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['label'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['label']=le.fit_transform(df['label'])
```

```python
df['label']
```

```python
X=df.drop('label',axis=1)
Y=df['label']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
reg=RandomForestClassifier(n_estimators=200)
reg.fit(X_train,Y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_pred,Y_test)
cm
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(y_pred,Y_test)
ac
```

```python
# prompt: take user input add that to the model and give me the predicted value

from sklearn.preprocessing import LabelEncoder
import numpy as np
import pandas as pd

def predict_crop(N, P, K, temp, humidity, ph, rain):
  """
  Predicts the recommended crop based on user input features.

  Args:
    N: Nitrogen content in the soil.
    P: Phosphorus content in the soil.
    K: Potassium content in the soil.
    temp: Temperature (Celsius).
    humidity: Humidity (percentage).
    ph: pH level of the soil.
    rain: Rainfall (mm).

  Returns:
    The predicted crop name.
  """
  # Create a DataFrame from the user input
  user_data = pd.DataFrame([[N, P, K, temp, humidity, ph, rain]], columns=X.columns)

  # Make the prediction
  prediction = reg.predict(user_data)

  # Inverse transform the prediction to get the crop name
  predicted_crop = le.inverse_transform(prediction)

  return predicted_crop[0]

# Get user input
N_input = float(input("Enter Nitrogen (N) content: "))
P_input = float(input("Enter Phosphorus (P) content: "))
K_input = float(input("Enter Potassium (K) content: "))
temp_input = float(input("Enter Temperature (Celsius): "))
humidity_input = float(input("Enter Humidity (%): "))
ph_input = float(input("Enter pH level: "))
rain_input = float(input("Enter Rainfall (mm): "))

# Get the prediction
predicted_crop_name = predict_crop(N_input, P_input, K_input, temp_input, humidity_input, ph_input, rain_input)

print(f"The recommended crop is: {predicted_crop_name}")
```
